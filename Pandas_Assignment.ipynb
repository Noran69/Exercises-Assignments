```python
import pandas as pd
# Load a dataset (e.g., Titanic, Iris, or any dataset of your choice) into a Pandas DataFrame.
df = pd.read_csv("/content/sample_data/titanic.csv")
df
```

```python
#Display the first 5 rows and the last 5 rows of the dataset.
df.head()
```

```python
df.tail()
```

```python
# Check the dimensions of the dataset (number of rows and columns).
df.shape

```

```python
# Display the column names and data types of each column.
df.info()
```

```python
# Check for missing values in the dataset.
df.isnull().sum()
```

```python
#Handle missing values by either filling them with a suitable value (e.g., mean, median, or mode) or dropping rows/columns with missing values.
df = df.fillna(0)
df
```

```python
#Remove duplicate rows from the dataset.
df = df.drop_duplicates()
df
```

```python
# Convert columns to appropriate data types (e.g., convert strings to datetime or categorical variables).
df['Age'] = df['Age'].replace('zero', 0).astype(int)
df
```

```python
# Rename columns to make them more descriptive.
df = df.rename(columns={'Pclass': 'Passenger Class'})
df
```

```python
# Drop unnecessary columns from the dataset.
df = df.drop(['Ticket'], axis=1)
df
```

```python
#Create a new column based on existing columns (e.g., calculate a derived metric).
df['Fare_Per_Person'] = df['Fare'] / df['Passenger Class']
df
```

```python
#Filter rows based on a condition (e.g., select rows where a column value is greater than a threshold).
df[df['Age'] > 18]
```

```python
# Sort the dataset by one or more columns.
df.sort_values(by='Age')
```

```python
# Group the dataset by a categorical column and calculate summary statistics (e.g., mean, median, count) for each group.
df.groupby('Sex')['Age'].mean()
```

```python
# Use the apply() function to apply a custom function to a column.
df['Fare'] = df['Fare'].apply(lambda x: x * 2)
df
```
