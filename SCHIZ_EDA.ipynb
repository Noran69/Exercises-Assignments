### Outlier Detection

```python
# Example: Outlier detection for Age
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
outliers_age = df[(df['Age'] < Q1 - 1.5*IQR) | (df['Age'] > Q3 + 1.5*IQR)]
print("Outliers in Age (by IQR rule):", outliers_age.shape[0])

# Boxplot for Age to visualize outliers
plt.figure(figsize=(4,2))
sns.boxplot(x=df['Age'], color='lightblue')
plt.title('Boxplot of Age')
plt.show()
```

### Univariate analysis

```python
# Histogram of Age
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### The Global Assessment of Functioning, or GAF, scale is used to rate how serious a mental illness may be. It measures how much a person's symptoms affect their day-to-day life on a scale of 0 to 100.

```python
# Histogram of GAF Score
plt.figure(figsize=(6,4))
sns.histplot(df['GAF_Score'], bins=10, kde=True, color='lightgreen')
plt.title('GAF Score Distribution')
plt.xlabel('GAF Score')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Positive Symptom Scores
plt.figure(figsize=(6,4))
sns.histplot(df['Positive_Symptom_Score'], bins=10, kde=True, color='salmon')
plt.title('Positive Symptom Score Distribution')
plt.xlabel('Positive Symptom Score')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Gender
plt.figure(figsize=(4,3))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Income Level
plt.figure(figsize=(4,3))
sns.countplot(x='Income_level', data=df, palette='pastel')
plt.title('Income Level Count')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.show()
```

Observations from Univariate Plots:
Age is fairly uniformly distributed between 18:0, with a slight peak around middle ages.
GAF Score (global functioning) ranges from 10:100, with many patients scoring around 50:80.
Positive Symptom Score ranges 0:100; many patients have higher symptom scores (patients tend to fill the higher range).
Gender is nearly balanced (~50% Female, 50% Male).
Income Level (coded 0/1/2) is roughly evenly split (each level about one-third of patients).
These plots help ensure no obvious data entry errors (e.g., negative ages) and give intuition on distributions.

### Bivariate Analysis

```python
# Correlation heatmap for numeric features
plt.figure(figsize=(15,10))
corr_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python

# Scatter plot: Age vs GAF Score, colored by Diagnosis
plt.figure(figsize=(15,10))
sns.scatterplot(x='Age', y='GAF_Score', hue='Diagnosis', data=df, palette='Set1')
plt.title('Age vs GAF Score by Diagnosis')
plt.xlabel('Age')
plt.ylabel('GAF Score')
plt.legend(title='Diagnosis')
plt.show()
```

```python
# Boxplot: GAF Score by Diagnosis
plt.figure(figsize=(6,4))
sns.boxplot(x='Diagnosis', y='GAF_Score', data=df, palette='pastel')
plt.title('GAF Score by Schizophrenia Diagnosis')
plt.xlabel('Diagnosis (No=Control, Yes=Patient)')
plt.ylabel('GAF Score')
plt.show()
```

Key Insights from Bivariate Plots:
The correlation heatmap shows a strong positive correlation between Diagnosis and Positive_Symptom_Score (~ 0.85) and between Diagnosis and Negative_Symptom_Score (~ 0.84). There is a strong negative correlation between GAF_Score and these symptom scores (higher symptoms  lower functioning).
The scatter plot (Age vs GAF) reveals that older patients do not necessarily have higher or lower GAF; there is wide variability.
The box plot clearly shows that patients (Diagnosis = Yes) have much lower GAF scores (median ~35) compared to controls (median ~80). Controls cluster with high GAF and lower symptom scores, while patients show the opposite.
These relationships confirm expected clinical patterns (patients with schizophrenia have more severe symptoms and lower functioning).
