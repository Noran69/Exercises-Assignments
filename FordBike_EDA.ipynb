```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['end_station_id','start_station_id','start_station_latitude', 'start_station_longitude','end_station_latitude', 'end_station_longitude'],axis=1,inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# Convert 'start_time' to datetime objects
df['start_time'] = pd.to_datetime(df['start_time'])

# Separate date and hour
df['start_date'] = df['start_time'].dt.date   # Extracts date only
df['start_hour'] = df['start_time'].dt.hour # Extracts hour only

# Display result
print(df[['start_date', 'start_hour']])
```

```python
# Convert 'end_time' to datetime objects
df['end_time'] = pd.to_datetime(df['end_time'])

# Separate date and hour
df['end_date'] = df['end_time'].dt.date   # Extracts date only
df['end_hour'] = df['end_time'].dt.strftime('%H:00:00')  # Extracts hour only

# # Display result
print(df[['end_date', 'end_hour']])
df
```

```python
df.drop(['start_time','end_time'],axis=1,inplace=True)
df.head()
```

```python
df['duration_sec']= df['duration_sec'] //  60
df['duration_sec']= df['duration_sec'] /  60
df.rename(columns={'duration_sec': 'duration_hour'}, inplace=True)
df.head()
```

```python
df['duration_hour']= df['duration_hour'].round(2)
df.head()
```

```python
df['member_gender'].fillna('Unknown', inplace=True)
df
```
